Compute the differention of $$f(x) = cos(x)$$ which is $$f'(x) = -sin(x)$$

```python
#solution
import numpy as np
import matplotlib.pyplot as plt
```

we will use forward difference to comupute $$f'(x) = (f(x_j-1)-f(x_j)/h$$


```python
#determine the step size
delta_x = 0.1
#define the compute grids
x = np.arange(0,2*np.pi,delta_x)
#compute the function cos(x)
y = np.cos(x)
#compute the forward differnces
forward_diff = np.diff(y)/delta_x
```

```python
print(x)
```

```python
x_diff = x[:-1:]
```

```python
x_diff
```

```python
#compute exact solution
exact_solution = -np.sin(x_diff)
```

```python
#plot solution
plt.figure(figsize=(12,8))
plt.plot(x_diff,forward_diff,"-",label= "Finite difference approximation")
plt.plot(x_diff,exact_solution,label="exact solution")
plt.legend()
```

```python
max_error = max(abs(exact_solution-forward_diff))
print(f"The maximum error :{max_error*100:.3f},%")
```

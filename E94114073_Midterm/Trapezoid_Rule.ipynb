```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def function(x):
    f = -1 * (x**2) + (6) * x + 56
    return f
```

```python
p1 = -2 
p2 = 11
n= 81
exact_value = 632.666
x = np.linspace(p1, p2, n)
f= function(x)
Integral_trapezoid = np.trapz(f, x)
error = exact_value - Integral_trapezoid
```

```python
print('Integral_trapezoid:', Integral_trapezoid)
print('eroor:',error)

plt.plot(x, function(x), linewidth = 2, c ='r')
for n in range(0, n-1):
    c = ['darkorange', 'forestgreen', 'royalblue']
    x_range = [x[n], x[n], x[n+1], x[n+1]] 
    y_range = [0, f[n], f[n+1], 0] 
    plt.fill(x_range, y_range, color=c[n%3], alpha = 0.5) 
plt.show()
```

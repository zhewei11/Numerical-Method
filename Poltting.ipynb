```python
#import mudulus
import numpy as np
import matplotlib.pyplot as plt

#Example 1
# plot sin(x) for 0< x <10 (rad)
x_min = 0
x_max = 2.*np.pi
num_of_points = 101
```

```python
x = np.linspace(x_min,x_max,num_of_points)
#print(x)
```

```python
y = np.sin(x)
#print(y)
```

```python
plt.plot(x,y)
```

```python
#Example 2
# plot sin(x), sin^2(x), sin(2x) in a figure with different labels and colors
# color code 'r', 'g', 'b','k'
# shape code 'o' for cicle, 's' squares, 'd' for diamond, '+' for cross
# assgin font size fontsize = 16
# you can creat legend
```

```python
y_1 = np.sin(x)**2
y_2 = np.sin(2*x)
```

```python
plt.plot(x,y,'r:',label='sin(x)')
plt.plot(x,y_1,'m-',label='sin^2(x)')
plt.plot(x,y_2,'b--',label='sin(2x)')
plt.xlabel('x',fontsize = 12)
plt.ylabel('Three different sin functions for 0 to 2Pi')
plt.title('Example for plotting several functions')
plt.legend()
plt.axis('equal')
```

```python
#Example 3 
#Create subplot for sin(x), sin^2(x) and sin(2x)
plt.figure(figsize=(5,5))

#sub plot 1, at the top sin(x)
plt.subplot(3,1,1)
plt.plot(x,y,'r:')
plt.ylabel('sin(x)')
plt.axis('scaled') #make your subplot look good

#sub plot 2, at the center sin^2(x)
plt.subplot(3,1,2)
plt.plot(x,y_1,'c-')
plt.ylabel('sin^2(x)')
plt.axis('scaled') #make your subplot look good
plt.ylim(-1,1)

#sub plot 3, at the bottom sin(2*x)
plt.subplot(3,1,3)
plt.plot(x,y_2,'b--')
plt.ylabel('sin(2x)')
plt.axis('scaled') #make your subplot look good
```

```python
#Example 4
#plot sin((x^2+y^2)^1/2)
x = np.arange(-5,5.5,0.5)
y = np.arange(-5,5.5,0.5)
```

```python
#create a mesh meshgrid
x_grid,y_grid =np.meshgrid(x,y)
```

```python
#z = sin((x^2+y^2)^1/2)
z = np.sin(np.sqrt(x_grid**2+y_grid**2))#np.sqrt finds square roots for you
```

```python
#let's create two plot
fig_1 = plt.figure(figsize=(15,5))
ax_1 = fig_1.add_subplot(1,2,1)
contour_1 = ax_1.pcolormesh(x_grid,y_grid,z,cmap='RdYlGn')
#another subplot
ax_2 = fig_1.add_subplot(1,2,2)
contour_2 = ax_2.contourf(x_grid,y_grid,z,cmap='RdYlGn')
```
